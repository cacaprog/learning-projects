```python
# import data from csv
import csv
from math import floor
```

```python
# letÂ´s create a list for each column of the csv file
ages = []
sexes = []
bmis = []
num_children = []
smokers = []
regions = []
insurance_charges = []
```

```python
def load_data(lst, file_name, col_name):
    # open the csv file
    with open(file_name) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            lst.append(row[col_name])
    return lst
```

```python
# Load the data into lists
load_data(ages, 'insurance.csv', 'age')
load_data(sexes, 'insurance.csv', 'sex')
load_data(bmis, 'insurance.csv', 'bmi')
load_data(num_children, 'insurance.csv', 'children')
load_data(smokers, 'insurance.csv', 'smoker')
load_data(regions, 'insurance.csv', 'region')
load_data(insurance_charges, 'insurance.csv', 'charges')
```

```python
print(len(ages), len(insurance_charges), len(sexes))
```

```python
# convert to correct types
ages = [int(age) for age in ages]
num_children = [int(children) for children in num_children]
insurance_charges = [float(charge) for charge in insurance_charges]
bmis = [float(bmi) for bmi in bmis]
```

```python
class MedInfo():
    def __init__(self, ages, sexes, bmis, num_children, smokers, regions, insurance_charges):
        self.ages = ages
        self.sexes = sexes
        self.bmis = bmis
        self.num_children = num_children
        self.smokers = smokers
        self.regions = regions
        self.insurance_charges =insurance_charges

      
    def avg_age(self):
        total_ages = 0
                
        for age in self.ages:
            total_ages += age
        return f"The average of all patients is {total_ages/len(self.ages): .0f}"

    def unique_regions(self):
        unique_regions = set(self.regions)
        return list(unique_regions)

    def sex_cost(self):
        men = 0
        women = 0
        total_man = 0
        total_woman = 0

        for i in range(len(self.sexes)):
            if self.sexes[i] == 'male':
                men += self.insurance_charges[i]
                total_man += 1
            else:
                women += self.insurance_charges[i]
                total_woman += 1
        
        return f"The average of insurance for men {men/total_man: .0f} dollars while for women {women/total_woman: .0f} dollars"

    def cost_region(self):
        region_ins = {'northeast': 0, 'southeast': 0, 'southwest': 0, 'northwest': 0}
        region_amount = {'northeast': 0, 'southeast': 0, 'southwest': 0, 'northwest': 0}
        region_avg_cost = {'northeast': 0, 'southeast': 0, 'southwest': 0, 'northwest': 0}

        for i in range(len(self.regions)):
            for k in region_ins.keys():
                if self.regions[i] == k:
                    region_ins[k] += 1
                    region_amount[k] = self.insurance_charges[i]

        for k in region_avg_cost.keys():
            region_avg_cost[k] = region_amount[k] / region_ins[k]
            print(f"The {k} region with {region_ins[k]} registered patients have an average insurance cost of {region_avg_cost[k]: 2f} dollars\n")
        return region_ins,region_amount,region_avg_cost

 
```

```python
patient_info = MedInfo(ages, sexes, bmis, num_children, smokers, regions, insurance_charges)
print(patient_info.unique_regions())
print(patient_info.avg_age())
print(patient_info.sex_cost())
print(patient_info.cost_region())
```
