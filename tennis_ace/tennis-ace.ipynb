### Linear Regression

Let´s build a single feature linear regression model. 

Only one feature to predict one outcome.

```python
X = tennis_df[['ServiceGamesPlayed']] # feature
y = tennis_df[['Wins']] # outcome
```

```python
# First, let´s split our data into training ans test datasets
from sklearn.model_selection import train_test_split
# 80% for training and 20% for test
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python

# Fit the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

# Score the model
print('The score for train is: ')
print(model.score(x_train, y_train))
print('The score for test is: ')
print(model.score(x_test, y_test))
```

Our model is doing a good job!

```python
# Find the predicted outcome
prediction = model.predict(y_test)
plt.scatter(x_test, prediction, alpha=0.5)
plt.scatter(X, y, alpha=0.5)
#plt.plot(x_test, prediction, alpha=0.5)
plt.show()
```

Create more models with one feature and compare which is the best model.

```python
def feature_outcome(f, o):
    X = tennis_df[[f]] # feature
    y = tennis_df[[o]] # outcome

    x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
    model = LinearRegression()
    model.fit(x_train, y_train)
    # print the model score to compare each of them
    return model.score(x_test, y_test)

feature_outcome('DoubleFaults', 'Wins')
```

```python
feature_outcome('BreakPointsFaced', 'Wins')
```

```python
feature_outcome('BreakPointsOpportunities', 'Wins')
```

```python
feature_outcome('Aces', 'Wins')
```

The best score for _Wins_ outcome is a model with _BreakPointsOpportunities_ feature :)

Let´s predict the best yearly earnings with two features.

```python
def more_features_outcome(x_a, x_b, y):
    X = tennis_df[[x_a, x_b]] # features
    y = tennis_df[[y]] # outcome

    # Split the train and test data
    x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
    model = LinearRegression()
    # Fit the model
    model.fit(x_train, y_train)

    # Model coeficient and score
    print(model.coef_) # get the heights for each feature  
    print(model.score(x_test, y_test))

    # Predict y_values (yearly winning)
    y_predicted = model.predict(x_test)
    print('The average earnings predicted is: ', y_predicted.mean())

    
```

```python
more_features_outcome('BreakPointsOpportunities', 'DoubleFaults', 'Winnings')
```

```python
more_features_outcome('ServiceGamesPlayed', 'Aces', 'Winnings')
```

```python
more_features_outcome('BreakPointsFaced', 'DoubleFaults', 'Winnings')
```

```python
# Test two of the highest features
more_features_outcome('DoubleFaults', 'BreakPointsOpportunities', 'Winnings')
```

Now, let´s try models with more than 4 features

```python
def five_features_outcome(x_a, x_b, x_c, x_d, x_e, y):
    X = tennis_df[[x_a, x_b, x_c, x_d, x_e]] # features
    y = tennis_df[[y]] # outcome

    # Split the train and test data
    x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
    model = LinearRegression()
    # Fit the model
    model.fit(x_train, y_train)

    # Model coefficient and score
    print(model.coef_)
    print(model.score(x_test, y_test))

    # Predict y_values
    y_predicted = model.predict(x_test)
    print('The average earnings predicted is: ', y_predicted.mean())
```

```python
five_features_outcome('ReturnGamesWon', 'BreakPointsOpportunities', 'ServiceGamesWon', 'BreakPointsSaved', 'Wins', 'Winnings')
```

```python
five_features_outcome('BreakPointsFaced', 'BreakPointsOpportunities', 'ServiceGamesWon', 'ReturnGamesWon', 'Losses', 'Winnings')
```

### Conclusion

If you are a tennis player and your goal is to earn more money, you need to perform well in different attributes (as a player and a team member).

If we put together 5 features: 'ReturnGamesWon', 'BreakPointsOpportunities', 'ServiceGamesWon', 'BreakPointsSaved' and 'Wins' to get outcome 'Winnings' (total winnings in 'USD$' in a year
), we could predict the average of USD$ 224675.82

But, not every feature contribute equaly.
**ServiceGamesWon** (total number of games where the player served and won), have a coefficient of **86194.60977785**, higher than **BreakPointsSaved** (% of the time the player was able to stop the receiver from winning service game when they had the chance) and **Wins** (number of matches won in a year), coefficient 5511.61584747 and 12798.36399854 respectivily.

The **ReturnGamesWon** (total number of games where the player’s opponent served and the player won) have a negative coefficient: -70777.7177545.
